Task Manager

````python
tasks = []

while True:
  print("\n 1. Add Task")
  print("2. View Task")
  print("3. Exit")
  print("4. Delete Task")

  for i in range(len(tasks)):
    print(f"{i+1}. {tasks[i]}")

  choice = input("Enter choice: ")

  if choice == '1':
    task = input("Enter Task: ")
    tasks.append(task)
  elif choice == '2':
    for t in tasks:
      print(t)
  elif choice == '3':
    break
  elif choice == "4":
    for i in range(len(tasks)):
        print(i + 1, tasks[i])

    task_num = int(input("Enter task number to delete: "))
    tasks.pop(task_num - 1)
  else:
    print("Invalid choice")

````
